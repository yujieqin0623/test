# 运算符

Python 中的运算符主要分为算术运算符、比较（关系）运算符、赋值运算符、逻辑运算符、位运算符、成员运算符和身份运算符共 7 大类，运算符之间也是由优先级的，下面我们就来进行简单介绍常用运算符。

## 算数运算符（+, -, *, /, //, %)

算术运算符：主要用于两个对象算数计算(加减乘除等运算)

![运算符](./resource/运算符数学运算符.png)

````python
# 两个数相乘
2 * 2
````

````python
# x除以y,得到的结果为浮点数
10 / 5
````

````python
# // 整除，返回商的整数部分（向下取整）
5 // 2
````

````python
# %求余数：使用 % ，得到结果为除法的余数，若余数为0则结果也为0
9 % 4
````

````python
# 求幂运算，使用两个乘号 ** 表示
2 ** 3
````

## 比较运算符(>, <, >=, <=, !=, ==)

- 注意他返回的值是逻辑值True或Flase
- 所有比较运算符返回1表示真，返回0表示假。这分别与特殊的变量True和False等价。
- 字符串直接使用比较运算符，会根据各个字符的编码值逐一进行比较 'abc'>'c'

![运算符](./resource/运算符比较运算符.png)

````python
# 比较对象是否相等！！！！！！！注意与一个等号的区分
1 == 2
````

````python
# 不等于，比较两个对象是否不相等
1 != 2
````

````python
# 返回x是否大于等于y
5 >= 3
````

## 逻辑运算符(and, or, not)

![运算符](./resource/运算符逻辑运算符.png)

- and	只要有一个运算数是False，结果就是False；只有所有的运算数都是True，结果才是True

````python
2 > 1 and 5 > 3
````

- or	只有所有的运算数都是False，结果才是False;只要有一个运算数是True，结果就是True

````python
3 > 5 or 4 < 2 or 10 < 3
````

## if 语句

***注意***
<br >使用条件语句不要少写了**冒号（:）**，而且必须是**英文版本**的。
<br >通过语句的缩进来判断语句体，缩进默认为**4个空格**，不要Tab和空格混用，否则会编译报错

请看以下代码示例：

````python
val = 100
if val >= 100:
    print('value is equal or greater than 100')
````

## else语句

else 为可选语句，当需要在条件不成立时执行内容则可以执行相关语句：

````python
val = 88
if val >= 100:
    print('value is equal or greater than 100')# 条件成立时执行的代码
else:
    print('value is equal or less than 100')# 条件不成立时执行的代码
````

## elif语句

如果需要检查多个条件，就可以使用elif，它是else if的简写
<br >一个if语句可以接多个elif代码块和一个else代码块，如果所有的elif条件均为False，则执行else代码块
<br >最后一个else可以不写

````python
val = 80
if val >= 100:
    print('value is equal or greater than 100')
elif val > 10:
    print('value is greater than 10 but less than 100')
else:
    print('value is equal or less than 10')
````

# 文件读写

## 文件打开与关闭

### 打开文件open

在python，使用open函数，可以打开一个已经存在的文件，或者创建一个新文件
<br >如果执行正常，open()函数返回1个文件对象，通过该文件对象可以对文件进行读写操作。如果指定文件不存在、访问权限不够、磁盘空间不足或其他原因导致创建文件对象失败则抛出异常。

**open参数介绍**
- open(文件路径，访问模式)
- file：用来指定打开的文件（不是文件的名字，而是文件的路径）
- mode：打开文件时的模式，默认时r表示只读，具体见下图
- encoding：文件的编码方式，有ASCII，UTF-8，GB2312，GBK等
- newline：换行符

![文件读写](./resource/文件读写访问模式.png)

````python
# r：只读模式，默认，打开文件以后，只能读取，不能写入。如果文件不存在，会报错
f = open('test.txt','r')
f
````

### 关闭文件colse

使用open()函数打开文件，当对文件内容操作完以后，一定要**关闭文件**对象，这样才能保证所做的任何修改都确实被保存到文件中!

````python
f.close() #关闭文件
````

## txt文件

### txt文件的写入

- write() ：把字符串写入文件
- writelines() ：把字符串序列写入文件

````python
# w：写入模式，打开文件以后只能写入，不能读取。如果文件存在，会覆盖文件；如果文件不存在，会创建文件
f = open('test.txt', 'w')
f.write('hello world, i am here!\n')
f.close()
# 运行之后会在当前脚本文件所在的路径中创建一个文件test.txt,并写入内容
````

````python
# a：追加模式，会在最后追加内容。如果文件不能存在，则会创建文件；如果文件存在，会追加
f = open('test.txt', 'a')
f.write('hello !\n')
f.close()
````

````python
# with 语句不需要colse()函数
with open('write_txt', 'w') as my_file:
    my_file.write('This is my first file that I wrote with Python.')
````

###  txt文件的读取

- read() 函数，该函数会会一次性读取文件的全部内容，但若文件过大，内存就爆了,如果有size，则读出sixe个字节
- readline() 每次读取一行内容
- readlines()readlines可以按照行的方式把整个文件中的内容进行一次性读取，并且返回的是一个列表，其中每一行为列表的一个元素

````python
f = open('test.txt', 'r')
content = f.read()  #read()将所有的数据都读取出来
print(content)
f.close()  #关闭文件
````

````python
with open('test.txt', 'r') as f:
    content = f.read(5)  # 最多读取5个数据
#     content = f.readline()  #只读取一行数据
#     content = f.readlines()  #将所有行的数据，保存到一个列表里
    print(content)
````

````python
with open('test.txt', 'r') as f:
    for line in f: #一行行读取
        print(line)
#line.strip()
````

## 注意事项

写入和读取使用相同的编码格式
 - encoding：文件的编码方式，有ASCII，UTF-8，GB2312，GBK等

````python
#  a：追加模式，会在最后追加内容。如果文件不能存在，则会创建文件；如果文件存在，会追加
with open('test中文.txt', 'w', encoding='utf8') as fp:
    fp.write('众能联合 !\n')
    print(fp)
````

````python
with open('test中文.txt', 'r', encoding='gbk') as fp:
    content = fp.readline()  #只读取一行数据
    print(content)
# 读入和写文件的编码不同，导致读取进来的数据乱码
````
